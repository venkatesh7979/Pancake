```python
import cv2
import time
import numpy as np
a=0
ar=[]
count=0
font = cv2.FONT_HERSHEY_COMPLEX
cap=cv2.VideoCapture('IMG_4393.MOV')
if cap.isOpened()==False:
    print('No file')
while cap.isOpened():
    ret,dosa=cap.read()
    if ret==True:
        time.sleep(1/60)
        img=cv2.medianBlur(dosa,35)
        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        ret, thresh = cv2.threshold(gray,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
        kernel = np.ones((3,3),np.uint8)
        opening = cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernel, iterations = 2)
        contours, hierarchy = cv2.findContours(opening.copy(), cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
        for i in range(len(contours)):
            if hierarchy[0][i][3] == 0:
                cv2.drawContours(dosa, contours, i, (255, 0, 0), 10)
        area=cv2.contourArea(contours[0])
        s=str(int(area))
        s='Area = '+s
        cv2.putText(dosa,s, (10,40), font, 1, (255, 255, 255))
        ar.append(area)
        a+=area
        count+=1
        cv2.imshow('frame',dosa)
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    else:
        break
print(a/count)
print(ar)
cap.release()
cv2.destroyAllWindows()
```
